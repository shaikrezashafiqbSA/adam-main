```python
# initialise

# Initialise LLM Handler

from IPython.display import Image, display
from settings import GEMINI_API_KEY
from llm_handler.GHandler import GHandler

g_handler = GHandler(GEMINI_API_KEY)


import numpy as np
import pandas as pd

import google.generativeai as genai
import google.ai.generativelanguage as glm

model = 'models/embedding-001'
genai.configure(api_key=GEMINI_API_KEY)

for m in genai.list_models():
  if 'embedContent' in m.supported_generation_methods:
    print(m.name)



# Clean and process database
from utils.pickle_helper import pickle_this

"""
Database search with Q&A

This is from an existing proprietary repository controlled by a halal-certified institution (eg: MUIS, JAKIM etc). The repository contains a list of halal approved ingredients. The user can ask questions about the halal status of the ingredients and the system will return the halal status of the ingredients.


Now that the embeddings are generated, let's create a Q&A system to search these documents. You will ask a question about hyperparameter tuning, create an embedding of the question, and compare it against the collection of embeddings in the dataframe.

The embedding of the question will be a vector (list of float values), which will be compared against the vector of the documents using the dot product. This vector returned from the API is already normalized. The dot product represents the similarity in direction between two vectors.

The values of the dot product can range between -1 and 1, inclusive. If the dot product between two vectors is 1, then the vectors are in the same direction. If the dot product value is 0, then these vectors are orthogonal, or unrelated, to each other. Lastly, if the dot product is -1, then the vectors point in the opposite direction and are not similar to each other.

Note, with the new embeddings model (embedding-001), specify the task type as QUERY for user query and DOCUMENT when embedding a document text.
"""
# reformat ingredients_df into columns Title and Text where Title is column iname and text is string combination of 'ecode + description + category' columns in 
# the format: "ecode: description (category)"
def reformat_ingredients_df():
  ingredient_df = pickle_this(data=None, pickle_name="ingredient_df",path="./database/halal_checker/")
  ecode_df = pickle_this(data=None, pickle_name="ecode_df", path="./database/halal_checker/")
  ingredient_df['category'] = None

  
  # now append ecode_df to ingredient_df|
  ingredients_df0 = pd.concat([ingredient_df, ecode_df], ignore_index=True)
  ingredients_df = ingredients_df0.copy()
  ingredients_df['Title'] = ingredients_df['iname'].copy()
  ingredients_df.rename(columns={'id_status': 'halal_status'}, inplace=True)
  """
  id_status is the halal status of the ingredient
  (1, 'Halal'),
  (2, 'Haram'),
  (3, 'Mushbooh'),
  (4, 'Depends');
  """
  # now rename ingredients_df column halal_status from id_status to halal_status (halal, haram, etc)
  ingredients_df['halal_status'] = ingredients_df['halal_status'].map({1: 'halal', 2: 'haram', 3: 'mushbooh', 4: 'depends'})
  ingredients_df['Text'] = ingredients_df.apply(lambda row: f"description: {row['desc']} halal_status: {row['halal_status']}", axis=1)
  # to be considered- how to handle text data

  # ingredients_df = ingredients_df[['Title', 'Text']]
  return ingredients_df
# Get the embeddings of each text and add to an embeddings column in the dataframe

def embed_fn(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             task_type="retrieval_document",
                             title=title)["embedding"]

# reformat ingredients_df into columns Title and Text where Title is column iname and text is string combination of 'ecode + description + category' columns
df = reformat_ingredients_df() 

# prompt user if want to embed or not, if not skip:
if input("Do you want to re-embed the data? (y/n): ") == "y":
  df['Embeddings'] = df.apply(lambda row: embed_fn(row['Title'], row['Text']), axis=1)
  pickle_this(data=df, pickle_name="embedded_halal_df", path="./database/halal_checker/")
else:
  df = pickle_this(pickle_name="embedded_halal_df", path="./database/halal_checker/")
```
